문제 2 (코드 작성)
- for 루프와 .items() 메소드를 사용하여, 주어진 사전(dictionary) my_scores의 모든 키(key)와 값(value)을 "과목명: 점수" 형태로 출력하는 코드를 작성하시오



```python
my_scores = {
    'Math': 95,
    'English': 80,
    'Science': 90
}

# 여기에 코드를 작성하세요
for key, value in my_scores.items():
  print(f"{key} : {value}")
# (출력 예시)
# Math : 95
# English : 80
# Science : 90
```

문제 3 (코드 작성)
- for 루프와 if-else 문을 사용하여 1부터 10까지의 숫자 중 3의 배수이면 "3의 배수"라고 출력하고, 아니면 "3의 배수 아님"이라고 숫자와 함께 출력하는 코드를 작성하시오.

```python
# 1부터 10까지의 숫자를 포함
for i in range(1, 11):
    # 여기에 if-else 조건문을 작성하세요
    if i % 3 == 0:
        print(f"{i} 3의 배수")
    else:
        print(f"{i} 3의 배수 아님")


# (출력 예시)
# 1 3의 배수 아님
# 2 3의 배수 아님
# 3 3의 배수
# 4 3의 배수 아님
# ...
# 9 3의 배수
# 10 3의 배수 아님
```

문제 4 (코드 작성)
- 두 개의 숫자 a와 b를 인자로 받아, $a^2 + b$ (a의 제곱 더하기 b)를 계산하여 반환(return)하는 함수 calc_func(a, b)를 정의하시오.

```python
# 여기에 함수를 정의하세요
def calc_func(a, b):
    return a**2 + b


# --- 테스트 코드 (수정 불필) ---
result = calc_func(5, 10)
print(f"5^2 + 10 = {result}")  # 예상 출력: 5^2 + 10 = 35
```

문제 5 (코드 작성)
- 'f 문자열의 상세 옵션'을 참고하여, f-string을 사용해 변수 val1과 val2를 아래의 요구사항에 맞게 포맷팅하여 str_result 변수에 할당하시오.

- 요구사항
 1. val1 (float): 소수점 이하 두 번째 자리까지만 표시
 2. val2 (int): 총 5자리로 표시하되, 빈자리는 0으로 채움

```python
val1 = 3.14159265
val2 = 42

str_result = f"val1 = {val1:.2f}, val2 = {val2:05d}" # 여기에 f-string을 작성하여 할당하세요

# --- 테스트 코드 (수정 불필) ---
print(str_result)
# 예상 출력: val1 = 3.14, val2 = 00042
```
